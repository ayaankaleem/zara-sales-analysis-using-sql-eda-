```python
import matplotlib.pyplot as plt
section_sales = df.groupby('section')['Sales Volume'].mean()

plt.bar(section_sales.index, section_sales.values)
plt.title("Section vs Sales Volume")
plt.show()
```

```python
prormrsection_count = df['section'].value_counts()

plt.figure(figsize=(6,6))

plt.pie(section_count.values,
        labels=section_count.index,
        autopct='%1.1f%%', startangle=90 )

plt.title("Zara Product Section Analysis")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(df['price'], df['Sales Volume'])

plt.xlabel("Price")
plt.ylabel("Sales Volume")
plt.title("Price vs Sales Volume")

plt.show()
```

```python
price_trend = df.groupby('season')['price'].mean()

plt.figure(figsize=(8,5))

plt.plot(price_trend.index,
         price_trend.values,
         marker='o' )

plt.xlabel("Season")
plt.ylabel("Average Price")
plt.title("Season-wise Average Price Trend")

plt.grid()

plt.show()
```

```python
promo_sales = df.groupby('Promotion')['Sales Volume'].mean()
plt.figure(figsize=(8,5))
plt.bar(promo_sales.index, promo_sales.values,color='r')
plt.title("Promotion vs Sales Volume")
plt.xlabel("Promotion")
plt.ylabel("Sales Volume")
plt.grid()
plt.show()
```
